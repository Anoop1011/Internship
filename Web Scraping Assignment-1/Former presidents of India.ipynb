```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
PAGE = requests.get("https://presidentofindia.nic.in/former-presidents.htm")
PAGE
```

```python
soup = BeautifulSoup(PAGE.content)
soup
```

```python
Name = soup.find('div', class_="presidentListing")
Name.text.split()[0:3]
```

```python
Term_of_office = soup.find('div', class_="presidentListing")
Term_of_office.text.split()[7:14]
```

```python
Names = []
for i in soup.find_all('div', class_="presidentListing"):
    Names.append((i.text).split()[0:3])
Names
```

```python
Term_of_office = []
for i in soup.find_all('div', class_="presidentListing"):
    Term_of_office.append((i.text).split()[7:14])
Term_of_office
```

```python
print(len(Names),len(Term_of_office))
```

```python
df = pd.DataFrame({'Names': Names,'Term_of_office': Term_of_office})
df
```
